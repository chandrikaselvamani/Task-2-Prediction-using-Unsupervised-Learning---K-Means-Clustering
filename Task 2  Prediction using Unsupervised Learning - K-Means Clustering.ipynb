#### TO CALCULATE THE OPTIMUM 'K' FOR CLUSTERING USING ELBOW METHOD

The elbow method  is used in determining the number of clusters in a data set. The method consists of plotting the explained variation as a function of the number of clusters, and picking the elbow of the curve as the number of clusters to use.

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
wss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit_transform(x)
    wss.append(kmeans.inertia_)
    
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,11),wss,'tab:cyan')
plt.xlabel("Numbers of Clusters")
plt.ylabel("Within-Cluster sum of squares")
plt.title("Elbow Method")

```

From the above plot we can able to identify the optimum K as 3, means number of clusters that can be formed from the given data is 3.

```python
km = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
km_new=km.fit_predict(x)
```

```python
km_new
```

```python
centroid=km.cluster_centers_
```

```python
centroid
```

#### PLOTTING THE CLUSTERS WITH RESPECT TO ITS CENTROID

```python
plt.figure(figsize=(10,10))
plt.scatter(x.iloc[km_new==0,0],x.iloc[km_new==0,2],s=75,c='tab:green',label='Iris-virginica')
plt.scatter(x.iloc[km_new==1,0],x.iloc[km_new==1,2],s=75,c='tab:blue',label='Iris-setosa')
plt.scatter(x.iloc[km_new==2,0],x.iloc[km_new==2,2],s=75,c='tab:purple',label='Iris-versicolor')
plt.scatter(centroid[:,0],centroid[:,2],s=100,c='black',label='Centroids')
plt.legend()
plt.title('Iris data Using K-Means Clustering')
```

In the above plot we can understand how the data is clustered based on the optimum K value with respect to its centroids.
